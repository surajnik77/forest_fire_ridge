#**Exploratory Data Analysis**

```python
#Drop day,month,year columns for not focusing on that
df1 = df.drop(['day','month','year'],axis=1)
```

```python
df1.head()
```

```python
#Check Categories in Classes
df1.Classes.value_counts()
```

```python
#Encoding the categories in Classes
df1['Classes']= np.where(df1['Classes'].str.contains('not fire'),0,1)
```

```python
df1.head(5)
```

```python
df1.tail(5)
```

```python
df1.Classes.value_counts()
```

```python
#Plot Density plot for all features
plt.style.use('seaborn')
df1.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#Pie Plot for percentages in classes
per = df1[["Classes"]].value_counts()
per.plot(kind = 'pie', fontsize = 12,labels=['Fire','Not fire'], explode = (0, 0.06), autopct = '%2.1f%%')
plt.xlabel('Classes', weight = "bold", color = "#2F0F5D", fontsize = 14, labelpad = 20)
plt.ylabel('Percentages_of_classes',weight='bold')
plt.title('Pie Plot for percentages in classes',weight='bold')
plt.axis('equal')
plt.legend(labels=['Fire','Not fire'],loc = "best")
plt.show()
```

#**Fire Analysis of 'Sidi-Bel Region' by Month**

```python
#Monthly Fire Analysis
dftemp = df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.xlabel("Month",weight='bold')
plt.ylabel("Number of Fires",weight='bold')
plt.title("Fire Analysis of 'Sidi-Bel Region'",weight='bold')
plt.legend()
```

#**Fire Analysis of 'Brjaia Region' by Month**

```python
#Monthly Fire Analysis
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.xlabel("Month",weight='bold')
plt.ylabel("Number of Fires",weight='bold')
plt.title("Fire Analysis of 'Brjaia Region'",weight='bold')
plt.legend()
```

**-Its observed that August and September had the most number of forest fires for both regions. and from the above plot of months,we can understand few things.**                                                                                                                                                  

-Most of the fires happened in August and very high fires happened in only 3 months - June,july,August.

-Less fire was on September.
